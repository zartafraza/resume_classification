```python
# Import necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import LatentDirichletAllocation

# Load toy text data
data = pd.DataFrame({'text': ['This is a document about cats.',
                              'This is a document about dogs.',
                              'This is a document about dogs and cats.',
                              'This is a document about birds.',
                              'This is a document about fish.',
                              'This is a document about dogs and birds.',
                              'This is a document about cats and birds.',
                              'This is a document about fish and birds.']})

# Convert text data to tf-idf vectors
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(data['text'])

# Use k-means clustering to group documents
kmeans = KMeans(n_clusters=3)
kmeans.fit(vectors)
kmeans_clusters = kmeans.predict(vectors)

# Use LDA to identify latent topics in the data
lda = LatentDirichletAllocation(n_components=3)
lda.fit(vectors)
lda_topics = lda.transform(vectors)
```
